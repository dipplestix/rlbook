```python
from dynaq import *
import matplotlib.pyplot as plt
%matplotlib inline
from collections import Counter
```

```python
class Grid:
    def __init__(self, w, h):
        self.w = w
        self.h = h
        self.actions = [0, 2, 4, 6]
        self.states = [tuple([i, j]) for j in range(self.h) for i in range(self.w)]
        self.starting_state = tuple([3, 0])
        self.terminal_states = [tuple([8, 5])]
        self.obstacles = [tuple(s) for s in [[x, 2] for x in range(1, 9)]]
    
    def update(self, ep):
        if ep > 100:
            self.obstacles = [tuple(s) for s in [[x, 2] for x in range(0, 8)]]

    def take_action(self, state, action):
        #locations are x, y
        x = state[0]
        y = state[1]
        if action == 0: # up
            x += 0
            y += 1
        if action == 2: # right
            x += 1
            y += 0
        if action == 4: # down
            x += 0
            y += -1
        if action == 6: # left
            x += -1
            y += 0
        if x>= self.w:
            x = self.w - 1
        if y >= self.h:
            y = self.h - 1
        if y < 0:
            y = 0
        if x < 0:
            x = 0
        if tuple([x, y]) in self.obstacles:
            x = state[0]
            y = state[1]
        r = 0
        if tuple([x, y]) in self.terminal_states:
            r += 1
        return tuple([x, y]), r

        
        
```

```python
policy = DynaQ(grid, gamma=.95, alpha=.1, eps=.1, n=10)
len_run = [0]*50
for j in range(100):
    for i in range(50):
        policy.run_episode()
        new_len = list(Counter(policy.ep_log).values())
    len_run = [len_run[i] + new_len[i] for i in range(len(len_run))]
len_run = [x/100 for x in len_run]
```
